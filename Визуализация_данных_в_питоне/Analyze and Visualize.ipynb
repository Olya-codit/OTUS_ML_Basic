# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
# Создание массива
array = np.zeros(10, dtype=int)
array[4] = 1

# Трансформация в двумерный массив
array_2d = array.reshape(2, 5)
print(array_2d)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
# Создание массива
array = np.arange(10, 50)

# Разворот массива
reversed_array = array[::-1]

# Нахождение четных элементов
even_elements = reversed_array[reversed_array % 2 == 0]
print(even_elements)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
# Создание массива
array_2d = np.arange(9).reshape(3, 3)
print(array_2d)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
# Создание массива
random_array = np.random.random((4, 3, 2))

# Нахождение минимума и максимума
min_value = random_array.min()
max_value = random_array.max()

print(f"Минимум: {min_value}, Максимум: {max_value}")
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
# Создание массивов
array1 = np.random.randint(0, 10, (6, 4))
array2 = np.random.randint(0, 10, (4, 3))

# Матричное умножение
result = np.dot(array1, array2)
print(result)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
# Создание массива
array = np.random.random((7, 7))

# Среднее и стандартное отклонение
mean_value = array.mean()
std_deviation = array.std()

# Нормализация массива
normalized_array = (array - mean_value) / std_deviation

print(f"Среднее: {mean_value}, Стандартное отклонение: {std_deviation}")
print("Нормализованный массив:")
print(normalized_array)
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
tips = sns.load_dataset("tips")
```

### Посмотрите на первые 5 строчек

```python
tips.head()
```

### Узнайте сколько всего строчек и колонок в данных

```python
# Форма датасета (строки, колонки)
print(f"Строк: {tips.shape[0]}, Колонок: {tips.shape[1]}")
```

### Проверьте есть ли пропуски в данных

```python
print(tips.isnull().sum())
```

### Посмотрите на распределение числовых признаков

```python
# Распределение числовых признаков
tips.describe()
```

```python
# Визуализация распределения
numerical_columns = tips.select_dtypes(include=["float", "int"]).columns
for col in numerical_columns:
    sns.histplot(tips[col], kde=True)
    plt.title(f"Распределение {col}")
    plt.show()
```

### Найдите максимальное значение 'total_bill'

```python
max_total_bill = tips["total_bill"].max()
print(f"Максимальное значение 'total_bill': {max_total_bill}")
```

### Найдите количество курящих людей

```python
smokers_count = tips[tips["smoker"] == "Yes"].shape[0]
print(f"Количество курящих людей: {smokers_count}")
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
average_total_bill_by_day = tips.groupby('day')['total_bill'].mean()
print(average_total_bill_by_day)
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
# Медианное значение 'total_bill'
median_total_bill = tips['total_bill'].median()

# Фильтрация строк и вычисление среднего 'tip' по полу
average_tip_by_sex = tips[tips['total_bill'] > median_total_bill].groupby('sex')['tip'].mean()
print(average_tip_by_sex)
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker_binary'] = tips['smoker'].apply(lambda x: 1 if x == 'Yes' else 0)
print(tips[['smoker', 'smoker_binary']].head())
```

# III. Visualization

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
# Scatterplot для Male
sns.scatterplot(data=tips[tips['sex'] == 'Male'], 
                x='total_bill', 
                y='tip', 
                hue='smoker', 
                palette='Set1')
plt.title("Scatterplot для Male")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.legend(title="Smoker")
plt.show()

# Scatterplot для Female
sns.scatterplot(data=tips[tips['sex'] == 'Female'], 
                x='total_bill', 
                y='tip', 
                hue='smoker', 
                palette='Set2')
plt.title("Scatterplot для Female")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.legend(title="Smoker")
plt.show()
```
